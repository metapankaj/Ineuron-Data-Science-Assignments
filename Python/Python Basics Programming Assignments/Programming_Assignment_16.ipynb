**Question1.**
<br>**Write a function that stutters a word as if someone is struggling to read it. The
first two letters are repeated twice with an ellipsis ... and space after each, and then the
word is pronounced with a question mark ?.**
<br>Examples
<br>stutter(&quot;incredible&quot;) ➞ &quot;in... in... incredible?&quot;
<br>stutter(&quot;enthusiastic&quot;) ➞ &quot;en... en... enthusiastic?&quot;
<br>stutter(&quot;outstanding&quot;) ➞ &quot;ou... ou... outstanding?&quot;

<br>Hint :- Assume all input is in lower case and at least two characters long.

```python
def stutter():
    word = input("Enter a word: ")
    return word[:2] + "... " + word[:2] + "... " + word + "?"

print(stutter())
```

**Question 2.**
<br>**Create a function that takes an angle in radians and returns the corresponding
angle in degrees rounded to one decimal place.**
<br>Examples
<br>radians_to_degrees(1) ➞ 57.3
<br>radians_to_degrees(20) ➞ 1145.9
<br>radians_to_degrees(50) ➞ 2864.8

```python
import math

def radians_to_degrees():
    radians = float(input("Enter an angle in radians: "))
    degrees = radians * (180 / math.pi)
    return round(degrees, 1)

print(radians_to_degrees())
```

**Question 3.**
<br><b>In this challenge, establish if a given integer num is a Curzon number. If 1 plus
2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon
number.
Given a non-negative integer num, implement a function that returns True if num is a Curzon
number, or False otherwise.</b>

<br>Examples
<br>is_curzon(5) ➞ True
<br># 2 ** 5 + 1 = 33
<br># 2 * 5 + 1 = 11
<br># 33 is a multiple of 11

<br>is_curzon(10) ➞ False
<br># 2 ** 10 + 1 = 1025
<br># 2 * 10 + 1 = 21
<br># 1025 is not a multiple of 21

<br>is_curzon(14) ➞ True
<br># 2 ** 14 + 1 = 16385
<br># 2 * 14 + 1 = 29
<br># 16385 is a multiple of 2


```python
def is_curzon():
    num = int(input("Enter a non-negative integer: "))
    numerator = 2 ** num + 1
    denominator = 2 * num + 1
    return numerator % denominator == 0

print(is_curzon())
```

**Question 4.**
<br>**Given the side length x find the area of a hexagon.**



<br>Examples
<br>area_of_hexagon(1) ➞ 2.6
<br>area_of_hexagon(2) ➞ 10.4
<br>area_of_hexagon(3) ➞ 23.4

```python
import math

def area_of_hexagon():
    x = float(input("Enter the side length of the hexagon: "))
    area = (3 * math.sqrt(3) * x ** 2) / 2
    return round(area, 1)

print(area_of_hexagon())
```

**Question 5.** 
<br>**Create a function that returns a base-2 (binary) representation of a base-10
(decimal) string number. To convert is simple: ((2) means base-2 and (10) means base-10)
010101001(2) = 1 + 8 + 32 + 128.
Going from right to left, the value of the most right bit is 1, now from that every bit to the left
will be x2 the value, value of an 8 bit binary numbers are (256, 128, 64, 32, 16, 8, 4, 2, 1).**
<br>Examples
<br> binary(1) ➞ &quot;1&quot;
<br> 1 * 1 = 1
<br> binary(5) ➞ &quot;101&quot;
<br> 1 * 1 + 1 * 4 = 5
<br>binary(10) ➞ &quot;1010&quot;
<br> 1 * 2 + 1 * 8 = 10

```python
def binary():
    decimal = int(input("Enter a decimal number: "))
    if decimal == 0:
        return "0"

    binary = ""
    while decimal > 0:
        remainder = decimal % 2
        binary = str(remainder) + binary
        decimal = decimal // 2

    return binary

print(binary())
```
